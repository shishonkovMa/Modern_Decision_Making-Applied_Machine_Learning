    Since we have the task of binary classification, it would be possible to use the "accuracy" metric, but this is incorrect for unbalanced data, because it is convenient for the algorithm to predict the majority class for all observations. Therefore, you can choose the measures "F-score", "Balanced accuracy", "AUC-PR" and "AUC-ROC". 

    However, in our case, the positive class (income of a person with more than 50k/year) is significantly smaller in size than the negative one, which means that AUC-ROC can give an inadequate assessment of the quality of the algorithm, since it measures the proportion of incorrectly accepted objects relative to the total number of negative ones. 

    The F-score does not take into account TN at the time. And we could stop at it, assuming that the model will predict the majority class (negative) well and we do not need to monitor its estimate. 

    But still I would like to get a complete picture, so we will opt for **balanced accuracy**. We will get the average of the accuracy of both classes.

# 2. Data preprocessing
We will conduct data preprocessing based on previously made conclusions.

```python
to_drop = ['AHRSPAY', 'AUNTYPE', 'CAPGAIN', 'CAPLOSS', 'DIVVAL', 'GRINREG', 'GRINST', 'PRCITSHP', 'VETQVA',
           'AMJIND', 'AMJOCC', 'HHDREL', 'MIGMTR3', 'MIGMTR4']

def df_replacement(df_in):
    df = df_in.copy()
    # AHGA
    map_AHGA = {' Children': 0,
                ' Less than 1st grade': 1,
                ' 1st 2nd 3rd or 4th grade': 2,
                ' 5th or 6th grade': 3,
                ' 7th and 8th grade': 4,
                ' 9th grade': 5,
                ' High school graduate': 6,
                ' 10th grade': 7,
                ' 11th grade': 8,
                ' 12th grade no diploma': 9,
                ' Some college but no degree': 10,
                ' Bachelors degree(BA AB BS)': 11,
                ' Masters degree(MA MS MEng MEd MSW MBA)': 12,
                ' Associates degree-occup /vocational': 13,
                ' Associates degree-academic program': 14,
                ' Prof school degree (MD DDS DVM LLB JD)': 15,
                ' Doctorate degree(PhD EdD)': 16}
    df.AHGA.replace(map_AHGA, inplace=True)

    # AHSCOL
    df.AHSCOL.replace({' Not in universe': 0, 
                          **{i: 1 for i in df.AHSCOL.value_counts().index if i != ' Not in universe'}}, 
                      inplace=True)
    # AMARITL
    df.AMARITL.replace({i: 1 if i == ' Married-A F spouse present' or i == ' Married-civilian spouse present'
                             else 0 for i in df.AMARITL.unique()},
                       inplace=True)
    # ARACE
    df.ARACE.replace({i: 1 if i == ' White' else 0 for i in df.ARACE.unique()},
                     inplace=True)
    # AREORGN
    df.AREORGN.replace({i: 1 if i == ' All other' else 0 for i in df.AREORGN.unique()},
                       inplace=True)
    # ASEX
    df.ASEX.replace({i: 1 if i == ' Male' else 0 for i in df.ASEX.unique()},
                     inplace=True)
    # HHDFMX
    df.HHDFMX.replace({df.HHDFMX.value_counts().index[ind]: ind+1 
                       if i >= 10000 else 0 for ind, i in enumerate(df.HHDFMX.value_counts())},
                      inplace=True)
    # PEFNTVTY, PEMNTVTY, PENATVTY
    pef = ['PEFNTVTY', 'PEMNTVTY', 'PENATVTY']
    for k in pef:
        df[k].replace({i: 1 if i == ' United-States' else 0 for i in df[k].unique()},
                      inplace=True)
    return df
        
train_repl = df_replacement(train)
```

## 2.2. Properly encode categorical features. Explain your choice of methods of categorical feature encoding

As I said above, we have only one oridnal feature - AHGA. And in order to avoid confusion with LabelEncoder, which incorrectly translates this categorical attribute into a numeric one, starting from 0. I worried about it in advance and determined the stages of education in ascending order from 0 to 16. Where 16 is PhD. With such layouts, we avoid using LabelEncoder, and classify this attribute as numeric.

Thus, we have all the features conditionally divided into 2 groups: nominal and numeric.

To encode nominal values, we will use CatBoostEncoder and standardize our data. And for numerical ones, we will only standardize.

Based on this approach, we will not produce a lot of new features by the OHE method, and we will also avoid leakage of the target.

```python
nominal_repl = [i for i in train_repl.columns if train_repl[i].dtype != 'int64' and i not in to_drop]
numeric_repl = [i for i in train_repl.columns if train_repl[i].dtype == 'int64' and i not in to_drop and i != 'target']

X_train_repl = train_repl.drop('target', axis=1)
y_train = train_repl['target']
X_test_repl = test_repl.drop('target', axis=1)
y_test = test_repl['target']


nominal_repl_pipe = make_pipeline(
    CatBoostEncoder(),
    StandardScaler(),
)

numeric_repl_pipe = make_pipeline(
    StandardScaler(),
)

column_transformer_repl = ColumnTransformer([
 ('nominal', nominal_repl_pipe, nominal_repl),
 ('numeric', numeric_repl_pipe, numeric_repl)
], remainder='drop')

column_transformer_repl.fit(X_train_repl, y_train)
X_train_repl_transf = column_transformer_repl.transform(X_train_repl)
X_test_repl_transf = column_transformer_repl.transform(X_test_repl)
```

In the case of the second option (where we use only categorical feature encoding and dimension reduction using PCA), we will use the same coding techniques, as well as standardize the features.

```python
category_pca = [i for i in train.columns if train[i].dtype != 'int64']
numeric_pca = [i for i in train.columns if train[i].dtype == 'int64' and i != 'target']

X_train_pca = train.drop('target', axis=1)
X_test_pca = test.drop('target', axis=1)


category_pca_pipe = make_pipeline(
    CatBoostEncoder(),
    StandardScaler(),
)

numeric_pca_pipe = make_pipeline(
    StandardScaler(),
)

column_transformer_pca = ColumnTransformer([
 ('nominal', category_pca_pipe, category_pca),
 ('numeric', numeric_pca_pipe, numeric_pca)
], remainder='drop')

column_transformer_pca.fit(X_train_pca, y_train)
X_train_pca_transf = column_transformer_pca.transform(X_train_pca)
X_test_pca_transf = column_transformer_pca.transform(X_test_pca)
```

## 3.1. Apply different techniques which can help to work with imbalanced data. You are supposed to try at least 2 techniques (in the end, you don't have to include them into your final pipeline - but you should explain why)

For a change, we will use 2 different techniques: undersampling (RandomUnderSampler) and oversampling (BorderlineSMOTE). But we will first study on a random forest model to assess the need to use unbalanced data techniques.

```python
def resampling(X_tr, y_tr):
    forest = RandomForestClassifier(max_depth=5, n_estimators=50, random_state=13)
    skf = StratifiedKFold(n_splits=5)
    
    sdf = [None, 
           BorderlineSMOTE(sampling_strategy=0.8, random_state=13),
           RandomUnderSampler(random_state=13)]
    
    name = ['Without changes']
    bas = []
    for ind, i in enumerate(sdf):
        ba = []
        if ind != 0:
            X_tr, y_tr = i.fit_resample(X_tr, y_tr)
            name.append(str(i).split('(')[0])
        for _, (train_index, test_index) in enumerate(skf.split(X_tr, y_tr)):
            forest.fit(X_tr.iloc[train_index], y_tr[train_index])
            ba.append(balanced_accuracy_score(forest.predict(X_tr.iloc[test_index]), y_tr[test_index]))
        bas.append(np.mean(ba))    

    plt.figure(figsize=(10,6))
    plt.plot(name, bas)
    plt.grid()
    plt.title('Balanced Accuracy')
    
resampling(X_train_repl_transf, y_train)
```

What we see in the end is that any of the techniques underestimates the result compared to the default RandomForest model. At the same time, adding observations of the minority class reduces the result more than removing observations of the majority class.

We can see the same story on the dataset that we have prepared for PCA.

```python
resampling(X_train_pca_transf, y_train)
```

**We will not use this approach.**

## 3.3. Perform blending or stacking of the trained machine learning algorithms. Carefully tune the weights of the ensemble

Let's create stacking from the best models (on the same data). Or rather it is: forest_pca, lgbm_pca, catboost_pca.

```python
# forest.best_estimator_.fit(feature_train, target_train)
# prediction = grid_pipe.best_estimator_.predict(feature_test)
# accuracy_score(target_test, prediction)

X_tr = X_train_repl_transf
y_tr = y_train
skf = StratifiedKFold(n_splits=5)

for _, (train_index, test_index) in enumerate(skf.split(X_tr, y_tr)):
    forest_pca.fit(X_tr.iloc[train_index], y_tr[train_index])
    lgbm_pca.fit(X_tr.iloc[train_index], y_tr[train_index])
    catboost_pca.fit(X_tr.iloc[train_index], y_tr[train_index])
    pred_a = forest_pca.predict(X_tr.iloc[test_index])
    pred_b = lgbm_pca.predict(X_tr.iloc[test_index])
    pred_c = catboost_pca.predict(X_tr.iloc[test_index])
    if _ == 0:
        stack = pd.DataFrame({'a': pred_a, 'b': pred_b, 'c': pred_c, 'y': y_tr[test_index]})
    else:
        sub_total_df = pd.DataFrame({'a': pred_a, 'b': pred_b, 'c': pred_c, 'y': y_tr[test_index]})
        stack = pd.concat([stack, sub_total_df], ignore_index=True)

X_stack = stack.drop('y', axis=1)
y_stack = stack.y

# create Meta algorithm with LogisticRegression
meta = LogisticRegression(penalty='l2', random_state=13)
meta.fit(X_stack, y_stack)
print('Mean Balanced Accuracy on STACKING with StratifiedKFold =', 
      balanced_accuracy_score(meta.predict(X_stack), y_stack))
```

```python
importances = forest.feature_importances_
feature_names = [i for i in X_train_repl_transf.columns]
std = np.std([tree.feature_importances_ for tree in forest.estimators_], axis=0)

forest_importances = pd.Series(importances, index=feature_names)

fig, ax = plt.subplots()
forest_importances.plot.bar(yerr=std, ax=ax)
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```
